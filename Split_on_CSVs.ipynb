```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('DATA/статистика_старт.xlsx - Данные.csv')
df = df.apply(lambda a: a.map(lambda b: np.nan if b in ['', '    ', '\t'] else b))
df
```

```python
df = df.drop_duplicates()
df = df.dropna(how='all')
df_change_columns = df.iloc[1:]
df_change_columns.columns = df.iloc[0:1].T[0]
df_change_columns
```

```python
specifity = ['Доходы гостиниц и аналогичных средств размещения',
             'Число гостиниц и средств размещения',
             'Численность размещенных лиц в гостиницах и ананлогичных средствах размещения',
             'Число ночевок в гостиницах и ананлогичных средствах размещения',
             'Средняя численность работников, включая внешних совместителей',
             'Число туристских фирм',
             'Средства, поступившие от реализации туристкого продукта',
             'Число реализованных турпакетов населению',
             'Индекс стоимости жизни']
```

```python
x = list()
for i in range(len(df)):
    if list(df['Unnamed: 0'])[i] in specifity:
        x.append(i)
x.append(len(df))
x
```

```python
dfs = dict()
for i in range(len(x)-1):
    k = df.iloc[x[i]:x[i+1]]
    k.columns = df_change_columns.columns
    f = k.iloc[1:].T
    f.columns = f.iloc[0:1].T[np.nan].to_list()
    g = f.iloc[1:].reset_index(drop=True)
    g['year'] = df_change_columns.columns[1:]
    g = g[g.columns.to_list()[-1:] + g.columns.to_list()[:-1]] 
    # .reset_index(drop=True)
    # print(f.iloc[1:].rename(index={np.nan: 'year'}).index)
    dfs[k[np.nan].to_list()[0]] = g #f.iloc[1:].rename(index={np.nan: 'year'})#, inplace=True)
dfs[list(dfs.keys())[0]]
```

```python
for i in dfs:
    dfs[i].to_csv(f'DATA/Split/{i}.csv')
```

```python
ter_exp = set()
for i in dfs:
    for j in dfs[i].columns:
        ter_exp.add(j)
ter_exp
```

```python
ter = dict()
for i in dfs:
    for j in dfs[i].columns[1:]:
        ter[' '.join(filter(lambda a: not a.isnumeric() and a != '', j.split(' ')))] = pd.DataFrame(columns=specifity)
ter
```

```python
for i in ter_exp:
    for j in dfs:
        if i in dfs[j].columns[1:]:
            # print(dfs[j][i].rename_axis('year'))
            ter[' '.join(filter(lambda a: not a.isnumeric() and a != '', i.split(' ')))]['year'] = dfs[j][dfs[j].columns[0]]
            ter[' '.join(filter(lambda a: not a.isnumeric() and a != '', i.split(' ')))][j] = dfs[j][i]
```

```python
ter[list(ter.keys())[0]]
```

```python
for i in ter:
    ter[i].to_csv(f'DATA/Ters/{i}.csv')
```
