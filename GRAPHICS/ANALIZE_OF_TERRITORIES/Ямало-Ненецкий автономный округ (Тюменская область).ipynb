# Импорт

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

# Загрузка Датасета

```python
data = pd.read_csv(
    '../../DATA/Ters/Ямало-Ненецкий автономный округ (Тюменская область).csv').drop('Unnamed: 0', axis=1)
```

```python
data.columns
```

# Предобработка

```python
data = data.apply(lambda s: s.map(lambda a: float(a.replace(',', '.')) if type(a) is str else a))
names = data.columns.to_numpy()[1:]
```

# Постройка граффиков год/параметр

```python
fig = plt.figure(tight_layout=True)
fig.set_figheight(25)
fig.set_figwidth(25)
gs = gridspec.GridSpec(3, 3)
n = 0
for i in range(3):
  for j in range(3):
    ax = fig.add_subplot(gs[i, j])
    ax.plot(data['year'], data[names[n]])
    ax.set_title(names[n])
    n+=1
```
