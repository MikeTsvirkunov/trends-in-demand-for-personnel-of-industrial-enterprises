# Импорт

```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import matplotlib.gridspec as gridspec
```

# Загрузка Датасета

```python
dfs_describe_ter = dict([(i, pd.read_csv(i).drop('Unnamed: 0', axis=1).apply(lambda s: s.map(lambda a: float(a.replace(',', '.')) if type(a) is str else a)))
                         for i in glob.glob("..\\..\\DATA\\Ters\\*")])
data = dict()
```

```python
for i in dfs_describe_ter:
    x, y = dfs_describe_ter[i].shape
    if dfs_describe_ter[i].isna().sum().sum() < 2/3 * x*y:
        data[i] = dfs_describe_ter[i].corr(method='pearson')
```

```python
data[list(data)[0]].loc['Доходы гостиниц и аналогичных средств размещения']['Доходы гостиниц и аналогичных средств размещения']
```

# Строим корреляции для параметров.

```python
corr_params = dict()
for i in data:
    corr_params[i] = dict()
    for col in data[i].columns:
        for row in data[i].index:
            if col != row and data[i].loc[col][row]>0.8:
                corr_params[i][(col, row)] = data[i].loc[col][row]
corr_params

                


```

# Считаем число датафреймов свойственных для корреляций

```python
num_of_corr_params = dict()
for i in corr_params:
    for j in list(corr_params[i]):
        z = tuple(sorted(j))
        if z in num_of_corr_params:
            num_of_corr_params[z]['num'] += 1
            num_of_corr_params[z]['corr_lvl'] += corr_params[i][j]
        else:
            num_of_corr_params[z] = dict()
            num_of_corr_params[z]['num'] = 1
            num_of_corr_params[z]['corr_lvl'] = corr_params[i][j]
num_of_corr_params
```

# Отфильтровываем коррелирующие параметры, что свойственны для большинства.

```python
for i in filter(lambda k: num_of_corr_params[k]['num'] > 8, num_of_corr_params):
    print(f'{" / ".join(i)}: {num_of_corr_params[i]["corr_lvl"] / num_of_corr_params[i]["num"]}')
```
