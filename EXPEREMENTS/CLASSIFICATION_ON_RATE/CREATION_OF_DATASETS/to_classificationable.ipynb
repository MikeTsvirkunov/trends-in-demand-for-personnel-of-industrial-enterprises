```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('../../../DATA/Concatination_of_ters/Without_Индекс стоимости жизни/normed_nan.csv').drop('Unnamed: 0', axis=1)
```

```python
c = [True]
for i in range(1, data.shape[0]):
    if data['Средняя численность работников, включая внешних совместителей'][i-1] < data['Средняя численность работников, включая внешних совместителей'][i]:
        c.append(True)
    else:
        c.append(False)
```

```python
data['Средняя численность работников, включая внешних совместителей'] = c
```

```python
data.dropna()
```

```python
data.to_csv('../../../DATA/Concatination_of_ters/Without_Индекс стоимости жизни/classes.csv')
```
